## Import Library

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from Config import Config
from Trainer import Trainer
from Predictor import Predictor
from Preprocessor import Preprocessor
```

## Data Load & EDA

```python
train = pd.read_csv(Config.data_path + Config.train_data, engine='python')
test = pd.read_csv(Config.data_path + Config.test_data, engine='python')
```

```python
train.info()   #  occyp_type null 값을 가짐
```

```python
## NA Imputation

occyp_null = train[train["occyp_type"].isnull()]
occyp_not_null = train[train["occyp_type"].notnull()]

plt.figure(figsize=(20,10))

plt.subplot(121)
plt.hist(occyp_null["income_total"])
plt.title('occyp_type_null')

plt.subplot(122)
plt.hist(occyp_not_null["income_total"])
plt.title('occyp_type_not_null')

#  occyp_type이 Na인 경우  소득이 현저히 낮음을 알 수 있고, 이로 인해 무직으로 추정  -> 특정  값을 넣기보단 "무직"이라는  하나의 카테고리를 생성
```

```python
## Log scale을 통해서 정규성을 높이고자 함

plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.hist(train[train["credit"] ==0]["income_total"])
plt.title('credit_0')
plt.subplot(2,2,2)
plt.hist(train[train["credit"] ==1]["income_total"])
plt.title('credit_1')
plt.subplot(2,2,3)
plt.hist(train[train["credit"] ==2]["income_total"])
plt.title('credit_2')

plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.hist(np.log(train[train["credit"] ==0]["income_total"]))
plt.title('credit_0_log')
plt.subplot(2,2,2)
plt.hist(np.log(train[train["credit"] ==1]["income_total"]))
plt.title('credit_1_log')
plt.subplot(2,2,3)
plt.hist(np.log(train[train["credit"] ==2]["income_total"]))
plt.title('credit_2_log')
```

```python
## 동일 고객의 Data가 상당수 존재 -> 각 고객의 고유 ID를 생성 
## 동일 고객이더라도 begin_month가 달라지는 것으로 보아 다른 시점의 Data가 각각 존재함을 알 수 있음

train[train[train.columns.drop("index")].duplicated(keep = False)].sort_values(["DAYS_BIRTH","income_total"])
```

```python
train['ID'] = \
    train['child_num'].astype(str) + '_' + train['income_total'].astype(str) + '_' +\
    train['DAYS_BIRTH'].astype(str) + '_' + train['DAYS_EMPLOYED'].astype(str) + '_' +\
    train['work_phone'].astype(str) + '_' + train['phone'].astype(str) + '_' +\
    train['email'].astype(str) + '_' + train['family_size'].astype(str) + '_' +\
    train['gender'].astype(str) + '_' + train['car'].astype(str) + '_' +\
    train['reality'].astype(str) + '_' + train['income_type'].astype(str) + '_' +\
    train['edu_type'].astype(str) + '_' + train['family_type'].astype(str) + '_' +\
    train['house_type'].astype(str) + '_' + train['occyp_type'].astype(str)
```
